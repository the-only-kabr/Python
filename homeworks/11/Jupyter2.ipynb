```python
# pip install sympy

from sympy import *
from sympy.plotting import plot
import matplotlib

```

```python
x = Symbol('x')
```

```python
y = -12*x**4*sin(cos(x)) - 18*x**3+5**2 + 10*x - 30
```

```python
plot(y,(x, -100, 100), show = True)
print(y)
```

```python
diff(y)
```

```python
plot(diff(y))
```

```python
x1, x2 = solve(y)
```

```python
type(solve(y))
```

```python
round(x1.evalf(),3)
```

```python
round(x2.evalf(),3)
```

```python
solve(diff(y))
```

```python
solve(y>0)
```

```python
solve(y<0)
```

```python
solve(diff(y)>0)
```

```python
solve(diff(y)<0)
```
